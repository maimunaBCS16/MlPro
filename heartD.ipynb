```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd

heart_data = pd.read_csv('/content/heart_dataset.csv')

```

```python
d=heart_data.drop(columns='target',axis=1)
h=heart_data['target']
print(d)
```

spilt test and train

```python
dtr,dte,htr,hte= train_test_split(d,h,test_size=0.2,stratify=h,random_state=2)
print(d.shape,dtr.shape,dte.shape)
```

train ML

```python
#logistic regression
model=LogisticRegression()
model.fit(dtr,htr)
```

```python
#score train
dtr_pred=model.predict(dtr)
trscore=accuracy_score(dtr_pred,htr)
print(trscore)

#score test
dte_pred=model.predict(dte)
tescore=accuracy_score(dte_pred,hte)
print(tescore)
```

build system

```python
#testing
insiddata=(56,0,1,140,294,0,0,153,0,1.3,1,0,2) #you can change these inputs
insiddata_as_numpy_array=np.asarray(insiddata)
insiddata_reshaped=insiddata_as_numpy_array.reshape(1,-1)
predict=model.predict(insiddata_reshaped)
print(predict)
if(predict[0]==0):
  print('healthy')
else:
  print('sick')
```
